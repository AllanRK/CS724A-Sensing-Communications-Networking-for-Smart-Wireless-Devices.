```python
#here we are considering the RSSI samples at 6 different locations.
# here we are considering the RSSI samples at 6 different smartphone orientations at each location.
# creating a list distance in order to store the distances.

def get_distance():
    dist = []
    for i in range(1,7):
        dist.append(i)
    return dist

distance = get_distance()


# to get the log values of the distances under consideration
def get_log_distance(dist):
    log_d = []
    for i in dist:
        x = math.log(i,10)
        log_d.append(x)
    return log_d

log_distance = get_log_distance(distance)
#print(log_distance)

# here s-s1, s_s2, s_s3, s_s4, s_s5 and s_s6 indicate signal strength in dBm
#at distances 1m , 2m, 3m , 4m , 5m and 6m respectively.
s_s1 = [-36,-35,-39,-40,-32,-37]  
s_s2 = [-59,-65,-59,-40,-55,-63]  
s_s3 = [-59,-68,-67,-75,-65,-72]   
s_s4 = [-75,-65,-66,-74,-76,-69]   
s_s5 = [-76,-79,-80,-82,-69,-72]  
s_s6 = [-86,-91,-89,-90,-87,-95] 

# calculating average signal strength in dBm at distances 1m , 2m, 3m , 4m, 5m and 6m respectively.

def get_avg_signal_strength(signal_strength):
    avg = sum(signal_strength)/6
    return avg

avg_signal_strength1 = get_avg_signal_strength(s_s1)
avg_signal_strength2 = get_avg_signal_strength(s_s2)
avg_signal_strength3 = get_avg_signal_strength(s_s3)
avg_signal_strength4 = get_avg_signal_strength(s_s4)
avg_signal_strength5 = get_avg_signal_strength(s_s5)
avg_signal_strength6 = get_avg_signal_strength(s_s6)

#displaying the average signal strength...
print("Average Signal Strength at distance of 1m in dBm is ",avg_signal_strength1)
print("Average Signal Strength at distance of 2m in dBm is ",avg_signal_strength2)
print("Average Signal Strength at distance of 3m in dBm is ",avg_signal_strength3)
print("Average Signal Strength at distance of 4m in dBm is ",avg_signal_strength4)
print("Average Signal Strength at distance of 5m in dBm is ",avg_signal_strength5)
print("Average Signal Strength at distance of 6m in dBm is ",avg_signal_strength6)

#list to store average signal strength...
list_signal_strength = []
list_signal_strength.append(avg_signal_strength1)
list_signal_strength.append(avg_signal_strength2)
list_signal_strength.append(avg_signal_strength3)
list_signal_strength.append(avg_signal_strength4)
list_signal_strength.append(avg_signal_strength5)
list_signal_strength.append(avg_signal_strength6)
#print(list_signal_strength)

    
```

Question 1-e Here we determine the variance of the RSSI samples w.r.t the best fit line

```python
#calculate the variance 
def get_variance(data0,data1):
    n = len(data0)
    deviations = []
    for i in range(n):
        z = math.pow((data0[i] - data1[i]),2)
        deviations.append(z)    
    var = sum(deviations) / n
    return var

#obtain the variance...
variance = get_variance(y,list_signal_strength)
print("Variance of RSSI samples w.r.t the best fit line: ",variance)
```
